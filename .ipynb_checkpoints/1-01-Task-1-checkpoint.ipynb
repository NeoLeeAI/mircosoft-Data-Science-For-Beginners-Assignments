## Goal
We will start with a text about Machine Learning, extract keywords from it, and then try to visualize the result.   


```python
url = 'https://en.wikipedia.org/wiki/Machine_learning'
```

## Step 1: Getting the Data 
First step in every data science process is getting the data. We will use requests library to do that

```python
import requests

text = requests.get(url).content.decode('utf-8')
print(text[:1000])
```

***
## Step 2: Transforming the Data
HTML code contains various tags and elements that are not relevant to your analysis. Therefore, it's necessary to remove these tags and extract only the text


```python
from html.parser import HTMLParser
class MyHTMLParser(HTMLParser):
    script = False
    res = ""
    def handle_starttag(self, tag, attrs):
        if tag.lower() in ["script","style"]:
            self.script = True
    def handle_endtag(self, tag):
        if tag.lower() in ["script","style"]:
            self.script = False
    def handle_data(self, data):
        if str.strip(data)=="" or self.script:
            return
        self.res += ' '+data.replace('[ edit ]','')

parser = MyHTMLParser()
parser.feed(text)
text = parser.res
print(text[:1000])
```

***
## Step 3: Getting Insights

### RAKE
   short for Rapid Automatic Keyword Extraction, is an easy-to-understand method for finding important words or phrases in a text. It helps to figure out the main ideas or subjects within the content.

In simple steps, RAKE does the following:

1.Splits the text into words and phrases, using common words like "and" or "the" as separators.  
2.Gives a score to each word and phrase based on how often they appear in the text.  
3.Sorts the words and phrases by their scores.   


RAKE is a fast and uncomplicated way to find the main ideas in a text, like in your project with the Wikipedia article on Machine Learning.


```python
import sys
!{sys.executable} -m pip install nlp_rake
```

```python
import nlp_rake

extractor = nlp_rake.Rake(max_words=2,min_freq=4,min_chars=5)
res = extractor.apply(text)
res
```

## Step 4: Visualizing the Result
We choose wordcloud to show the result. Wordcloud is a suitable choice for visualization in my project for several reasons: 1. Intuitive representation 2. Easy to understand

```python
import matplotlib.pyplot as plt
def plot(pair_list):
    k,v = zip(*pair_list)
    plt.bar(range(len(k)),v)
    plt.xticks(range(len(k)),k,rotation='vertical')
    plt.show()

plot(res)
```

```python
!{sys.executable} -m pip install wordcloud
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(background_color='white',width=800,height=600)
plt.figure(figsize=(15,7))
plt.imshow(wc.generate_from_frequencies({ k:v for k,v in res }))
```

```python
plt.figure(figsize=(15,7))
plt.imshow(wc.generate(text))
```
